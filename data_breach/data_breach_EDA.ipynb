# Data Breaches

source for the dataset: https://en.wikipedia.org/wiki/List_of_data_breaches

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv("dataBreacheslist.csv")
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['Method'] = data['Method'].str.lower()
```

```python
data.sort_values(by='Year').head()
```

```python
data['Method'].unique()
```

```python
data['Organization type'].unique()
```

```python
data['Year'].unique()
```

```python
# Hacker method every year
data.groupby(['Year','Method'])['Organization type'].agg(["count"]).reset_index()
```

```python
# hacker method
dd = data.groupby(['Method'])['Entity'].agg(["count"]).reset_index()
sns.set({'figure.figsize':(18,8)})

chart=sns.barplot(data=dd, x='Method', y='count')
chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
None
```

```python
# organization hacked group by year
data.groupby(['Year','Organization type'])['Organization type'].agg(["count"])
```

```python
for x in range(len(data.groupby(['Entity'])['Method'].count())):
    if data.groupby(['Entity'])['Method'].count().values[x] > 1:
        print(data.groupby(['Entity'])['Method'].count().index[x]," --> data breach ",data.groupby(['Entity'])['Method'].count().values[x]," times")
```
